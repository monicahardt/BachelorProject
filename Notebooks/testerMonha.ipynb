```python
import matplotlib.pyplot as plt
import numpy as np

def plot_word2vec_matrices():
    fig, ax = plt.subplots(2, 3, figsize=(15, 8))
    
    # Vocabulary size (V) and Embedding size (N)
    V, N = 10, 5

    # One-hot encoded vector for a target word (e.g., "fox")
    one_hot = np.zeros(V)
    target_word_index = 3  # Index of "fox"
    one_hot[target_word_index] = 1

    # Weight matrices
    W = np.random.rand(V, N)
    W_prime = np.random.rand(N, V)

    # Embedding extraction (one-hot * W)
    embedding = one_hot @ W

    # Prediction (embedding * W')
    prediction = embedding @ W_prime

    # Plot one-hot encoding
    ax[0, 0].imshow(one_hot.reshape(V, 1), cmap='Blues')
    ax[0, 0].set_title('One-hot Encoding (Input)')
    ax[0, 0].axis('off')

    # Plot input-to-hidden weights W
    ax[0, 1].imshow(W, cmap='Greens')
    ax[0, 1].set_title('Weight Matrix W (Input to Hidden)')
    ax[0, 1].axis('off')

    # Plot hidden layer (embedding)
    ax[0, 2].imshow(embedding.reshape(1, N), cmap='Reds')
    ax[0, 2].set_title('Hidden Layer (Embedding)')
    ax[0, 2].axis('off')

    # Plot hidden-to-output weights W'
    ax[1, 0].imshow(W_prime, cmap='Purples')
    ax[1, 0].set_title("Weight Matrix W' (Hidden to Output)")
    ax[1, 0].axis('off')

    # Plot output layer (before softmax)
    ax[1, 1].imshow(prediction.reshape(1, V), cmap='Oranges')
    ax[1, 1].set_title('Output Layer (Prediction)')
    ax[1, 1].axis('off')

    # Plot softmax probabilities
    softmax = np.exp(prediction) / np.sum(np.exp(prediction))
    ax[1, 2].imshow(softmax.reshape(1, V), cmap='Greys')
    ax[1, 2].set_title('Softmax Probabilities')
    ax[1, 2].axis('off')

    plt.tight_layout()
    plt.show()

plot_word2vec_matrices()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_word2vec_cbow_example():
    fig, ax = plt.subplots(2, 3, figsize=(15, 8))
    
    # Vocabulary size (V) and Embedding size (N)
    V, N = 4, 10  # Smaller vocabulary and higher dimensions

    # One-hot encoded vectors for context words (e.g., "cat" and "dog")
    one_hot_cat = np.zeros(V)
    one_hot_dog = np.zeros(V)
    one_hot_cat[0] = 1  # "cat"
    one_hot_dog[1] = 1  # "dog"

    # Average one-hot vectors
    one_hot_avg = (one_hot_cat + one_hot_dog) / 2

    # Weight matrices
    W = np.random.rand(V, N)
    W_prime = np.random.rand(N, V)

    # Embedding extraction (one-hot_avg * W)
    embedding = one_hot_avg @ W

    # Prediction (embedding * W')
    prediction = embedding @ W_prime

    # Plot one-hot encoding of "cat" and "dog"
    ax[0, 0].imshow(np.vstack((one_hot_cat, one_hot_dog)).reshape(V, 2), cmap='Blues')
    ax[0, 0].set_title('One-hot Encoding (Context)')
    ax[0, 0].axis('off')

    # Plot average of one-hot encodings
    ax[0, 1].imshow(one_hot_avg.reshape(V, 1), cmap='Greens')
    ax[0, 1].set_title('Averaged One-hot Encoding')
    ax[0, 1].axis('off')

    # Plot input-to-hidden weights W
    ax[0, 2].imshow(W, cmap='Reds')
    ax[0, 2].set_title('Weight Matrix W (Input to Hidden)')
    ax[0, 2].axis('off')

    # Plot hidden layer (embedding)
    ax[1, 0].imshow(embedding.reshape(1, N), cmap='Purples')
    ax[1, 0].set_title('Hidden Layer (Embedding)')
    ax[1, 0].axis('off')

    # Plot hidden-to-output weights W'
    ax[1, 1].imshow(W_prime, cmap='Oranges')
    ax[1, 1].set_title("Weight Matrix W' (Hidden to Output)")
    ax[1, 1].axis('off')

    # Plot output layer (before softmax)
    ax[1, 2].imshow(prediction.reshape(1, V), cmap='Greys')
    ax[1, 2].set_title('Output Layer (Prediction)')
    ax[1, 2].axis('off')

    plt.tight_layout()
    plt.show()

plot_word2vec_cbow_example()
```
