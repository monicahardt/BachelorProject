# Visualisation of Loss during training of model

The model used for the project is either a model with a low epoch count or a model trainged over 2000 generations and then the model with the lowest loss were choosen. 



```python
# Open the file in read mode
with open('../Data/loss_tracker_100d_10000e.txt', 'r') as f:
    # Read the lines of the file into a list
    lines = f.readlines()

# Strip newline characters and convert strings to floats
lst = [float(line.strip()) for line in lines]
```

```python
# Open the file in read mode
with open('loss_tracker_2.txt', 'r') as f:
    # Read the lines of the file into a list
    lines = f.readlines()

# Strip newline characters and convert strings to floats
lst_2 = [float(line.strip()) for line in lines]
```

```python
lst = lst[50:]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Your list of values
your_list = lst

# Create x-axis values (assuming sequential indices)
x_values = range(1, len(your_list) + 1)

# Fit a linear regression model
coefficients = np.polyfit(x_values, your_list, 1)
regression_line = np.polyval(coefficients, x_values)

# Plot the data and regression line
plt.plot(x_values, your_list, linestyle='-', linewidth=2, label='Data')
plt.plot(x_values, regression_line, linestyle='--', color='red', label='Regression')

# Add labels, title, and legend
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss over the duration of 10000 -the first 50 epochs')
plt.legend()

# Show the plot
plt.show()
```
